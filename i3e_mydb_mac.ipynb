```python
"""
Extract all the company name of the mac we collected
"""
import sqlite3

PATH="data/WiGLE_Backup.db"

I3E="i3e"
NET = "network"
VEN = "vendor"

con = sqlite3.connect(PATH)
cur=con.cursor()
```

```python
query = "select bssid from {}".format(NET)
cur.execute(query)

sql="SELECT network.bssid,i3e.* FROM i3e, network where i3e.Assignment COLLATE NOCASE = SUBSTR(network.bssid, 1, 8)"

cur.execute(sql)

data = cur.fetchall()
print(len(data))
print(data)

#cur.execute("DELETE FROM vendor")
for row in data:
    cur.execute('INSERT OR IGNORE INTO vendor (mac,Registry,Assignment,OrganizationName,OrganizationAddress) VALUES (?,?,?,?,?)', row)
con.commit()
```
