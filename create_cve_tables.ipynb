```python
import sqlite3

PATH="data/WiGLE_Backup.db"

# Connect to the database
conn = sqlite3.connect(PATH)
c = conn.cursor()

# Create the cve_basic table
c.execute('''CREATE TABLE "cve_basic" (
        "id" TEXT,
        "keywords" TEXT,
        "sourceIdentifier" TEXT,
        "published" TEXT,
        "lastModified" TEXT,
        "vulnStatus" TEXT,
        "description_value" TEXT,
        PRIMARY KEY("id","description_value","keywords"));''')

# Create the cve_metrics table
c.execute('''CREATE TABLE cve_metrics_2
             (id TEXT,
             cvss_type TEXT,
              source TEXT,
              type TEXT,
              cvssVersion TEXT,
              vectorString TEXT,
              accessVector TEXT,
              accessComplexity TEXT,
              authentication TEXT,
              confidentialityImpact TEXT,
              integrityImpact TEXT,
              availabilityImpact TEXT,
              baseScore REAL,
              baseSeverity TEXT,
              exploitabilityScore REAL,
              impactScore REAL,
              acInsufInfo INTEGER,
              obtainAllPrivilege INTEGER,
              obtainUserPrivilege INTEGER,
              obtainOtherPrivilege INTEGER,
              userInteractionRequired INTEGER,
              PRIMARY KEY("id","type"))''')

c.execute('''CREATE TABLE cve_metrics_3
             (id TEXT,
             cvss_type TEXT,
              source TEXT,
              type TEXT,
              cvssVersion TEXT,
              vectorString TEXT,
              attackVector TEXT,
              attackComplexity TEXT,
              privilegesRequired TEXT,
              userInteraction TEXT,
              scope TEXT,
              confidentialityImpact TEXT,
              integrityImpact TEXT,
              availabilityImpact TEXT,
              baseScore REAL,
              baseSeverity TEXT,
              exploitabilityScore REAL,
              impactScore REAL,
              PRIMARY KEY("id","type"))''')
# Create the weakness table
c.execute('''CREATE TABLE cve_weaknesses
             (id TEXT,
              type TEXT,
              description_value TEXT,
              PRIMARY KEY("id","description_value"))''')

# Create the configuration table
# c.execute('''CREATE TABLE cve_configuration
#              (id TEXT PRIMARY KEY,
#               operator TEXT,
#               negate INTEGER,
#               vulnerable INTEGER,
#               criteria TEXT,
#               matchCriteriaId TEXT)''')

# Insert data into the tables
conn.commit()
conn.close()
```
